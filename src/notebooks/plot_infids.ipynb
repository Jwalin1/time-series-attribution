```python
import json
import matplotlib.pyplot as plt
```

```python
with open('captum_metrics.json', 'r') as fp:
    data = json.load(fp)    
```

```python
for dataset, methods in data.items():
  vals = [float(val) for val in methods.values()]
  plt.bar(range(len(vals)), vals)
  plt.xticks(range(len(vals)), list(methods.keys()), rotation=90)
  plt.xlabel('methods')
  plt.ylabel('infidelity')
  plt.title(dataset)
  plt.show()
```
