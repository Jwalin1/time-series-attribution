```python
captum_methods = ["Saliency", "IntegratedGradients", "InputXGradient", "GuidedBackprop", "LayerGradCam", "GuidedGradCam", "Lime"]
yiskw713_methods = ["GradCAMpp", "SmoothGradCAMpp", "ScoreCAM", "RISE"]
```

```python
%cd /content/time-series-attribution/src

import torch
import numpy as np
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
from modules import data_f, network_f, network_architectures, attribution_f

import warnings
warnings.filterwarnings("ignore")

dataset = "SyntheticAnomaly"
train_inputs, train_labels, test_inputs, test_labels = data_f.getRead_data(dataset)
x = train_inputs[train_labels==1][0]
plt.plot(np.transpose(x))
plt.show()
```

## Cascading randomization

```python
%cd /content/time-series-attribution/


methods = captum_methods + yiskw713_methods
rand_layers = range(0,-9,-1)
model,criterion,optimizer,scheduler = network_f.setupModel(network_architectures.AlexNet(2, 3))
network_f.load_state_dict(model, dataset)

fig, axs = plt.subplots(len(methods), len(rand_layers), figsize=(15,15))
fig.set_tight_layout(True)

for i,method in enumerate(tqdm(methods)): 
  for j,layer in enumerate(rand_layers):
    rand_model = network_f.randomize_layers(model, layer, "upto")
    map_ = attribution_f.applyMethodSample(method, rand_model, np.expand_dims(x,0))[0]
    axs[i][j].plot(np.transpose(map_))
    axs[i][j].xaxis.set_ticklabels([])
    axs[i][j].yaxis.set_ticklabels([])

    if i == len(methods)-1:
      layer_name = ""
      if j != 0:
        layer_name = str(network_f.get_children(model)[layer]).partition('(')[0]
      axs[len(methods)-1, j].set_xlabel(f'{layer} {layer_name}')
    if j == 0:
      axs[i, 0].set_ylabel(method)
```
